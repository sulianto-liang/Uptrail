```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sales_data.csv')
```

```python
df
```

DATA CLEANSING & FILL MISSING VALUES

```python
# remove NaN column ('Unnamed: 11' until 'Unnamed: 22')
df = df.dropna(axis='columns',how='all')
```

```python
# remove NaN row (row with all NaN data, e.g. row with index 20)
df = df.dropna(how='all')
```

```python
# remove duplicate rows
df = df.drop_duplicates()
```

```python
# get to know  of each column
df.info()
```

```python
# Change data type of column Date into datetime
df['Date']= pd.to_datetime(df['Date'])
```

```python
df
```

```python
# export cleaned data set into csv
df.to_csv('output.csv', index=False)
```

COLUMN ADDITION

```python
# Fill out Total_Amount column with Quantity * Price since there is Total_Amount data with NaN value
df['Total_Amount'] = df['Quantity'] * df['Price']
```

```python
# Add column Adv_to_Amount_% as Advertising_Spend to Total_Amount percentage
df['Adv_to_Amount_%'] = ((df['Advertising_Spend'] / df['Total_Amount']) * 100).round(2)
```

```python
# Add column Month_Year
df['MonthYear'] = df['Date'].dt.month.astype(str) + "-" + df['Date'].dt.year.astype(str)
```

```python
# Add column Net_Amount (Total_Amount substracted by Advertising_Spend)
df['Net_Amount'] = df['Total_Amount'] - df['Advertising_Spend']
```

```python
# Add column Profit_% as Net Amount to Total_Amount percentage
df['Profit_%'] = ((df['Net_Amount'] / df['Total_Amount']) * 100).round(2)
```
